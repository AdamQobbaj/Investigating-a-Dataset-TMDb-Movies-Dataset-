### Research Question 1: How are the budgets and revenues for movies related?

First thing that might come to mind when you are asked this question about this data set might be "Does the movie with highest budget have the gighest revenue?", therefore, we need to sort the rows accodring to the budget to answer basic questions. First, Let's see the biggest revenue for a movie ever.

```python
#Let's see the movie with the biggest revenue ever
df[df["revenue"] == df.revenue.max()]
```

Woah! Thats a huge revenue! And This budget does not seem that big compared to the revenue? Then let's check out the movie with the biggest budget.

```python
df[df["budget"] == df.budget.max()]
```

Aw... What a shame, a movie with this big of a budget has this little revenue? It must have been very hard on the producers! Anyway, comapred to the one with the biggest revenue, this movie's budget is almost twice the amount!

But those are simple queries about movies' budget and revenue,  let's dive deeper into statistics related to them.  

```python
# Making a copy of the dataframe with rows sorted according to budget in decending order
df_q1 = df.sort_values(by='budget', ascending=False)
df_q1.head()
```

To make it easy for us, we will divide movies to groups based on budget.

```python
# Defining the budget ranges
budget_bins = [0, 1000, 10000, 50000, 100000, 200000, 500000, 700000, 
               1000000, 2000000, 5000000, 7000000, 10000000, 20000000, 30000000, 
               40000000, 50000000, 70000000, 80000000, 100000000, 200000000, 500000000]

# Defining the corresponding budget range labels
budget_labels = [1000, 10000, 50000, 100000, 200000, 500000, 700000, 
               1000000, 2000000, 5000000, 7000000, 10000000, 20000000, 30000000, 
               40000000, 50000000, 70000000, 80000000, 100000000, 200000000, 500000000]

# Creating the new column 'budget_range_under'
df_q1['budget_range_under'] = pd.cut(df_q1['budget'], bins=budget_bins, labels=budget_labels, right=False)
df_q1.head()
```

```python
# Converting the 'budget_range_under' column to a categorical data type
df_q1['budget_range_under'] = pd.Categorical(df_q1['budget_range_under'], categories=budget_labels, ordered=True)
df_q1['budget_range_under'].dtype
```

```python
# Let's now divide the movies to groups based on budget range 
grouped_df = df_q1.groupby('budget_range_under').agg(
    total_revenue=('revenue', 'sum'),
    average_revenue=('revenue', 'mean')
).reset_index()
grouped_df.head(30)
```

That should make it easier to handle. Ugh, numbers are cool and all but I can't seem to comprehend the relation by just looking at them. I think a couple graphs can simplify things, no? Let's see...

```python
import matplotlib.pyplot as plt

# Plotting average profit against budget range
plt.figure(figsize=(10, 6))
plt.bar(grouped_df['budget_range_under'].astype(str), grouped_df['average_revenue'], color='lightblue')

# Adding labels and title
plt.xlabel('Budget Range')
plt.ylabel('Average Revenue')
plt.title('Average Revenue per Budget Range')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()

# Showing the plot
plt.show()
```

This plot alone should be enough to understan the relation between the budget and revenue, but i feel a pie chart could be helpful too...

```python
# Plotting average profit as a pie chart
plt.figure(figsize=(8, 8))
plt.pie(grouped_df['average_revenue'], labels=grouped_df['budget_range_under'].astype(str),
        autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)

# Adding title
plt.title('Average Profit per Budget Range')

# Showing the plot
plt.show()
```

```python
# Adding a column for the profit
df_q1['profit'] = df_q1['revenue'] - df_q1['budget']

# Anothr one for negative profit
df_q1['negative_profit'] = (df_q1['profit'] < 0)

# Grouping rows based on budget range, but this time, we are concerned with profit
grouped_df_negative = df_q1.groupby('budget_range_under').agg(
    negative_profit_sum=('negative_profit', 'sum')  # Sum of True values (negative profits count)
).reset_index()

grouped_df_negative.head(30)
```

We can see that most movie recieved good profit, but still, some seem to have lost profit. 
Let's plot them for better understanding.

```python
# Plotting average profit against budget range
plt.figure(figsize=(10, 6))
plt.bar(grouped_df_negative['budget_range_under'].astype(str), grouped_df_negative['negative_profit_sum'], color='lightblue')

# Adding labels and title
plt.xlabel('Budget Range')
plt.ylabel('Number of Movies')
plt.title('Number of Movies that Lost Profit Per Budget Range')
plt.xticks(rotation=45, ha='right')  # Rotate x-axis labels for better readability
plt.tight_layout()

# Showing the plot
plt.show()
```

We can see from the figure that the number of movies with negative profit seems to be higher for movies within ranges above 1,000,000 and below 70,000,000, that is, around the mid point. It seems like that if the budget was too large, the chance of faluire will be smaller. Same goes for smaller budget, that is, if your budget is too low, even if your didn't make a huge hit, a little revenue is enough to make up for the loses used as budget.

Now that we have seen how budget can affect revenue in different budget ranges, it pique my curiosity: What are the top 10 movies in terms of percent profit? Let's find this out:

```python
# Some movies' budget is still not very accurate, and may have been entered wrong (usually under 1000),
# so we will not include these movies. 
df_q1 = df_q1[df_q1['budget'] > 1000]

# Adding a row called percent_profit
df_q1['percent_profit'] = (df_q1['profit'] / df_q1['budget']) * 100
    
# Sorting according to percent profit
df_q1 = df_q1.sort_values(by='percent_profit', ascending=False)
df_q1.head(10)
```

Woah! I didn't expect that, the profit for the movie "Paranormal Activity" is almost 1288940% the budget, that's a huge number, even the 10th place is about 2323%, the movie business might be a great one. Also, in a way, this might actually be an answer for out main question "How are revenues and budgets related?", we can see that a movie with just a budget of only 15000 brought a huge profit with such low budget, but this is just an instance, and not representative of the avergae case.

### Research Question 2: Which Genres Need The Highest Budget or Have the Highest Revenue Rates?

After having discussed movies and thier budgets, revenues and profit, I became curiose about geners, and how they can affect budget and profit. A good question, isn't it? Surely different genres need different amount of budget, right? And it's not wrong to assume different genres have different popularity, and therefore revenues, is it? If this was true, the following analysis would confirm.

Before anything, why not check out the genres of the movie with the biggest budget? Let's see...

```python
df[df['budget'] == df['budget'].max()]['genres']
```

Intresting, but also to be expected. In any case, let's now check the genres of the movie with the highest revenue:

```python
df[df['revenue'] == df['revenue'].max()]['genres']
```

Hmm... We can see that 'Action' and 'Adventure' appeared in both, is there a connection? Let's analyze further:

```python
# Let's start by making a copy of the cleaned dataframe of the movies dataset
df_q2 = df.copy()

# This dataframe contains a column called 'genres' which has multiple values eperated by '|', we need
# to split each row based genre
df_q2['genres'] = df_q2['genres'].str.split('|')
df_q2 = df_q2.explode('genres', ignore_index=True)

# Renaming the column to 'genre', as each row now has one 'genre'
df_q2 = df_q2.rename(columns = {'genres' : 'genre'})
df_q2.head(10)
```

```python
# Converting the 'genre' column to a categorical data type
df_q2['genre'] = pd.Categorical(df_q2['genre'], ordered=True)
df_q2['genre'].dtype
```

```python
# Groups the dataframe by `group_col` and computes the average for columns in `value_cols`.
# Returns the resulting dataframe.
def group_and_average(df, group_col, value_cols):
    grouped_df = df.groupby(group_col)[value_cols].mean().reset_index()
    grouped_df.rename(columns={col: f'avg_{col}' for col in value_cols}, inplace=True)
    return grouped_df

# Plots a pie chart given a dataframe, value column, label column, and chart title.
def plot_pie_chart(df, value_col, label_col, title):
    plt.figure(figsize=(8,8))
    plt.pie(df[value_col], labels=df[label_col], autopct='%1.1f%%', startangle=140)
    plt.title(title)
    plt.axis('equal')
    plt.show()
```

```python
grouped_df_q2 = df_q2.groupby('genre')[['popularity', 'budget', 'revenue']].mean().reset_index()
grouped_df_q2.rename(columns={
    'popularity': 'avg_popularity',
    'budget': 'avg_budget',
    'revenue': 'avg_revenue'
}, inplace=True)
grouped_df_q2
```

Let's sort them by budget to see how expensive each genre can be. The top five genres in terms of budget are 

```python
grouped_df_q2.sort_values(by='avg_budget', ascending=False).head(5)
```

, which is to be expected, as all these seem to need expensive equipment and CGI to produce, and also high pay for the actors. Let's now see some charts:

```python
# Plotting average budget as a pie chart
plot_pie_chart(grouped_df_q2, 'avg_budget', 'genre', 'Average Budget by Genre')
```

But what about revenue? Are these (or at least some of these) genres going to appear in the top five list? Let's see...

```python
grouped_df_q2.sort_values(by='avg_revenue', ascending=False).head(5)
```

Oh? Four of the genres appeared again but with different order? They seem to be popular that many people watch them therefore give more revenue, don't they? Speaking of popularity, we will check that in the next section but now let's present the statistics visually, shall we?

```python
# Plotting average revenue as a pie chart
plot_pie_chart(grouped_df_q2, 'avg_revenue', 'genre', 'Average Budget by Genre')
```

Lastly, Why not check popularity? Top five popular genres:

```python
grouped_df_q2.sort_values(by='avg_popularity', ascending=False).head(5)
```

Still not far from the ones before huh? Okay, now comes the chart:

```python
# Plotting average popularity as a pie chart
plot_pie_chart(grouped_df_q2, 'avg_popularity', 'genre', 'Average Budget by Genre')
```
